(2) 아래 (가)와 (나)에 대해 옳은 발표를 하고 있는 인간들을 고르세요.

<img src="https://i.imgur.com/mYnIG85.png"></img>

```
갑: (가)는 변수로 선언한 이후에는 값을 바꿀 수 없어요. immutable 하다고도 하죠.
을: (가)는 순서가 있는 자료형이에요.
병: (나)는 API 의 반환 형태로 흔히 쓰이는 자료형이기도 합니다.
정: (나)의 "값" 필드에는 튜플 자료형이 올 수 없다는 걸 항상 유의해야 합니다.
```

**----<보기>----**<br>
```
    a. 갑, 을
    b. 갑, 병
    c. 을, 병
    d. 을, 정
    e. 병, 정
```

(3) 아래 코드의 출력 결과는?

```
test = {}
test[1] = 1
test['1'] = 2
test[1] += 1

sum = 0
for k in test:
    sum += test[k]
    
print(sum)
```


**----<보기>----**<br>
```
    a. 0
    b. 1
    c. 2
    d. 3
    e. 4
```
