````python
import json, requests

def get_book_info(book_title, sort_by='accuracy'):
    API_URL = 'https://dapi.kakao.com/v2/search/book'
    API_KEY = '9af2b40c42092a9bb605b3779eaad645'
    
    params = {
        'query' : book_title,          
        'sort' : sort_by,
    }
    
    headers = {
        'Authorization' : 'KakaoAK ' + API_KEY
    }
    
    response = requests.get(API_URL, headers=headers, params=params)
    json_data = response.json()
    
    return json_data

book_dict = get_book_info("에고라는 적")
print(book_dict)
````

## 먼저 ```while``` 문을 복습합시다
**"반복해서 문장을 수행해야 할 경우 while문을 사용한다. 그래서 while문을 반복문이라고도 부른다."** (점프 투 파이썬)


#### 예제 코드
* 보리차 식히기<br>

````python
temp = 100 # 지금 차의 온도

while temp > 90:
    print("현재 온도:", temp)
    temp -= 1

print("차가 다 식었습니다. 마셔도 좋습니다.")
````

````python
'''
에뛰드 하우스 시뮬레이션
'''

print("어서 오세요. 공주님 ^-^ 에뛰드 하우스입니다")

options = """
<아래 보기 중에서 하고 싶은 걸 선택하세요.>
1. 신상이 뭔지 물어본다.
2. 갑질을 한다.
3. 나간다.
"""

while True:
    print(options)
    choice = input()
    
    if (choice == '1'):
        print("이번 달 신상은 디어 달링 워터젤 틴트 아이스크림이에요")
    elif (choice == '2'):
        print("흑흑 저한테 왜 이러세요")
    elif (choice == '3'):
        print("안녕히 가세요 고갱님")
        break
    else:
        print("잘못 선택하셨습니다. 숫자 1, 2, 3 중 하나를 선택해서 입력하세요.")
````


## 반복문 2 - ```for```

"`for` loops are traditionally used when you have a block of code which you want to repeat a fixed number of times. <br>
**The Python `for` statement iterates over the members of a sequence in order, executing the block each time."** (Python Wiki)

#### 기본 구조
```python
for 변수 in 리스트(또는 튜플, 문자열):
    수행할 문장1
    수행할 문장2
    ...
```

#### 예제 코드
* "총 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 보여주시오."


````python
marks = [90, 25, 67, 45, 80]

number = 0 
for mark in marks: 
    number = number +1 
    if mark >= 60: 
        print("%d번 학생은 합격입니다." % number)
    else: 
        print("%d번 학생은 불합격입니다." % number)
````

3. 결과값이 없는 함수
